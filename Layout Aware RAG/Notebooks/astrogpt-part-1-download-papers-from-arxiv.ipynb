# Create a Dataframe for keeping track of the files

```python
dir_path = "./AstroPhysics_papers/*.pdf"


all_files = []




for file_path in glob.glob(dir_path):
    
    elem = {"filepath":file_path,
            "filename":file_path.split("/")[-1]
           }
    
    all_files.append(elem)
    
    

filepath_df = pd.DataFrame(all_files)

filepath_df.shape 
```

# Trash

```python
import subprocess

def run_bash_command(command:str,
                      ):

  resp = subprocess.run(
             command,
             text=True,
             stdout=subprocess.PIPE,
             check=True)


  return resp



def get_doc_paths(resp):
  all_paths = []
  for file in resp.stdout.split('\n'):
    all_paths.append(file)


  return all_paths
```
